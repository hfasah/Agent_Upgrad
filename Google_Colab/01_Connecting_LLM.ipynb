# 01. Connecting to OpenAI LLM

This notebook demonstrates how to connect to OpenAI's API and make a simple request using the Responses API.

```python
# Install OpenAI package
!pip install openai
```

```python
# Import required libraries
from openai import OpenAI
import json
```

```python
# 🔐 ENTER YOUR API KEY HERE
OPENAI_API_KEY = "YOUR_API_KEY_HERE"

# Initialize OpenAI client
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
# Make a request to OpenAI
response = client.responses.create(
    model="gpt-4o-mini",
    input="Tell me Upgrad inception story in 100 words.",
    instructions="Respond in a funny way and in Captain Jack Sparrow Style.",
    temperature=1,
    max_output_tokens=200,
)

print("Response from LLM:", response.output_text)
```
