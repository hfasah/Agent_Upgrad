This notebook demonstrates how to use the code interpreter tool for calculations and visualizations.

```python
from openai import OpenAI
import os
```

```python
# 🔐 ENTER YOUR API KEY HERE
OPENAI_API_KEY = "YOUR_API_KEY_HERE"

client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
print("🔧 Tool Demo: Code Interpreter")
print("=" * 60)

# Example 1: Simple calculation with visualization
print("\n📊 Example 1: Investment calculation with bar graph")

response = client.responses.create(
    model="gpt-5-nano",
    tools=[
        {"type": "code_interpreter", "container": {"type": "auto"}}
    ],
    input="I have invested USD 1000 at the rate of 10% per annum simple interest for 10 years. What will be the total amount after 10 years? Give me visualization of bar graph as well.",
)

print("✅ Result:")
print(response.output_text)
```
